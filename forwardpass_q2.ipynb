```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

# Try changing input here:
x = np.array([[0.0, 1.0]])  # try: [0,0], [0,1], [1,0], [1,1]

# Weights and biases (kept simple and small)
W1 = np.array([[0.4, -0.2],
               [0.1,  0.3]])
b1 = np.array([[0.0, 0.0]])

W2 = np.array([[0.2],
               [-0.5]])
b2 = np.array([[0.0]])

# ---- Forward pass ----
z1 = x @ W1 + b1
a1 = sigmoid(z1)            # change to relu(z1) and observe

z2 = a1 @ W2 + b2
y_hat = sigmoid(z2)

print("Input x:", x)
print("Hidden output a1:", np.round(a1, 4))
print("Final prediction y_hat:", np.round(y_hat, 4))
```
