```python
import math

# 1) Inputs
x1, x2 = 0.7, -0.3

# 2) Weights + bias (try changing these values)
w1, w2 = 0.5, -0.8
b = 0.2

# 3) Activations
def step(z, threshold=0.0):
    return 1.0 if z >= threshold else 0.0

def sigmoid(z):
    return 1.0 / (1.0 + math.exp(-z))

def relu(z):
    return max(0.0, z)

# 4) Forward propagation for one neuron
z = (x1 * w1) + (x2 * w2) + b

print("Net input (z):", round(z, 4))
print("Step output:", step(z))
print("Sigmoid output:", round(sigmoid(z), 4))
print("ReLU output:", round(relu(z), 4))
```
